## Question 1: 
*Rappeler et distinguer les notions d'indépendance et de non corrélation pour deux signaux aléatoires*

- Les deux signaux aléatoires s1(t,w) et s2(t,w) sont indépendants si la connaissance de la réalisation de l'un signal n'a aucune incidence sur la probabilité de réalisation de l'autre signal. Mathématiquement, on peut décrire ce charactère en utilisant la fonction densité spectrale du puissance comme ci-dessous:

$$ \Psi_{s1,s2}(w) = \Psi_{s1}(w)*\Psi_{s2}(w) $$

- La corrélation entre deux signaux aléatoires quant à elle envoie à l'existence d'une liaison linéaire entre les deux signaux aléatoires. Elle est mathématiquement mesuré  en calculant la fonction de l'intercorrélation:

$$\Gamma_{s_1,s_2}(t_1,t_2) = E[s_1(t_1)s_2(t_2)^*]$$

- Il est théoriquement démonstré que si dex signaux aléatoire indépendants alors ils ne sont pas corrélés.Par conséquence, si deux signaux aléatoires sont corrélés alors ils sont nécessairement dépendants. Enfin, il est possible deux signaux aléatoires non corrélés mais ils sont dépendants.

Implémenter la fonction qui blanchit les signaux entrées

```python
def blanchiment_des_signaux(x):
    
    """
    Cette function blanchit des singaux entrées
    
    Paramètres
    ----------
    
    x -  un vecteur répresant tous les signaux enregistrés.
        x.shape = (n,1,m)
            n: nombre du signaux blanchies observés
            m: nombre du samplings du chaque signal
    
    Sorites
    -------
    x_bl - signaux blanchies.
    
    """
    
    m = x.shape[2]
    n = x.shape[0]
    
    x_t = np.transpose(x,(1,0,2))
    
    x_mul = np.zeros((n,n,m))
    
    for i in range(m):
        x_mul[:,:,i] = np.dot(x[:,:,i],x_t[:,:,i])
    
    np.shape(x_mul)
    # la variance des singaux
    E_x = np.mean(x_mul, axis = -1)
    
    lamda, E = LA.eig(E_x)
    
    D = np.diag(np.power(lamda,-1/2))
    
    # les signaux blanchis
    matrice_blanchi = np.dot(E,np.dot(np.power(D,1),E.T))
    
    x_blanchi = np.zeros((n,1,m))
    
    for i in range(m):
        x_blanchi[:,:,i] = np.dot(matrice_blanchi,x[:,:,i])
    
    return x_blanchi
```

Implémenter la methode servant à trouver un vecteur w qui aide à trouver un seule composant indépendant.

```python
def usc(x):
    
    """
    Cette function sert à trouver un vecteur w.
    
    Paramètres
    ----------
    
    x : les signaux observés blanchies.
        x.shape = (n,1,m)
            n: nombre du signaux blanchies observés
            m: nombre des échantillons de chaque signal
    
    Sorties
    -------
    
    + vecteur w.
    
    """
    
    # les entrées de l'algorithme sont les signaux observés blanchies 
    x_blanchi = x
    
    # Initialiser le vecteur w.
    n = x_blanchi.shape[0]
    m = x_blanchi.shape[2]
    w = np.random.rand(n,1)
    #print('w.shape: ',w.shape) = (2, 1)
    
    epoch = 0
    # la methode d'itération de Newton.
    while True:
            
        # calcule le vecteur y
        y = np.zeros((1,1,m))
        #print('y.shape: ',y.shape) =  (1, 1, 220500)
        
        for i in range(m):
            #print('x_blanchi[:,:,i].shape: ',x_blanchi[:,:,i].shape) = (2, 1)
            y[:,:,i] = np.dot(w.T,x_blanchi[:,:,i])
            
        w_i = w
        w_plus = np.mean(np.multiply(x_blanchi,np.tanh(y)), axis = -1) - np.mean(np.power(np.cosh(y),-2),axis = -1)*w
        
        w = np.divide(w_plus,LA.norm(w_plus))
        
        if epoch > 30:
            break
        
        #else:
        #    if (epoch % 10) == 0:
        #        print('epoch: ',epoch,' cost -: ',LA.norm(w-w_i),' cost +: ',LA.norm(w+w_i))

        epoch = epoch + 1
        
    return w
```

Implémenter la fonction qui évalue les performances des reconstructions des signaux originaux.

```python
def EQMNi(s_chapeau, s):
    
    """
    Cette fonction evalue les eurreurs des reconstructions des signaux des résources
    
    paramètres
    ----------
    
    s_chapeau: signal de la reconstruction
        s_chapeau.shape = (m,1)
        
    s: signal de résource
        s.shape = (m,1)
    
    sorties
    -------
    
    E: erreur critique
    
    """
    
    s_chapeau_norm = np.linalg.norm(s_chapeau)
    s_t = np.transpose(s)
    s_t_norm = np.linalg.norm(s_t)
    s_st = np.dot(s_t,s_chapeau)
    E = 1 - (s_st/(s_chapeau_norm*s_t_norm))**2
    
    return E
```

Implémenter la méthode qui trouve tous les composantes indépendantes.

```python
def tlci(x):
    
    """
    Cette fonction sert à trover tous les composantes indépendantes.
    
    Paramètres
    ----------
    
    x : les signaux enregistrés.
        x.shape = (n,1,m)
            n: nombre du signaux blanchies observés.
            m: nombre du samplings du chaque signal
    
    Sorties
    -------
    
    n composants idépendants.
    
    """

    n = x.shape[0]
    
    W = np.zeros((n,n))
    
    # la première composante
    w1 = usc(x)
    
    W[0,:] = np.transpose(w1).ravel()
    
    k = 1 # le nombre des composantes cherchées
    
    while True:
        
        if k == n:
            break
        
        w_next = usc(x)
        
        répète = False # si wi est l'un des elements dans la matrice W
        
        for i in range(k):
            
            if np.array_equal(w_next.T,W[k,:]):
                répète = True
        
        w_ = np.zeros((n,1))
        
        if not répète:
            
            for i in range(k):
                w_ = w_ + np.dot(w_next.T,np.transpose([W[i,:]]))*(np.transpose([W[i,:]]))
                
            w_next = w_next - w_
            
            w_next = np.divide(w_next,LA.norm(w_next))
            
            W[k,:] = np.transpose(w_next).ravel()
            
            k = k +1 
    
    return W
```

Implementer la methode qui calcule la matrice d'autocorrélation entre deux vecteurs observés blanchies X1, X2

```python
def intercovariance(X1,X2, N, step):
    """
    
    La fonction calcule la matrice d'autocorrélation deux vecteurs observés blanchies X1, X2.
    
    Paramètres:
    ----------
    
    + X1,X2: les vecteurs observés blanchies sous la taille (M,1).
    + N: le nombre des echantillons extraits des vecteurs des observations X1,X2.
    + step: 
    
    Sorties:
    --------
    + R est la vecteur d'intercovariance entre X1, X2.
    """
    
    R = np.zeros((N,N))
    X = np.zeros((N,N))
    # nombre de échantillons totals
    M = X1.shape[0]
    
    # nombre de fois pour un boucle
    n = 0
    
    # x = X[i:j,1]
    i = 0
    
    while True:
        
        j = i + N
        
        if j > M:
            break
            
        x1 = X1[i:j,0];
        x1 = x1.reshape(x1.shape[0],1)
        x2 = X2[i:j,0];
        x2 = x2.reshape(x2.shape[0],1)
        
        x = np.dot(x1,x2.T)
        
        X = X + x
        
        n = n+1;
        
        i = i + step
    
    R =  X/n
    
    return R
```

Implémenter la methode off(.)

```python
def off(R):
    """
    Calculer l'opérateur off(.)

    """
    
    N = R.shape[0]
    
    F = (np.sum(R) - np.trace(R))/(N*(N-1))
    
    return F
```

Implémenter la methode tr(.)

```python
def tr(R):
    
    """
    Calculer l'opérateur tr(.)
    
    """
    
    N = R.shape[0]
    
    T = np.trace(R)/N
    
    return T
```

Implémenter une fonction qui cacule la matrice A

```python
def AFTabc(X1,X2,N,step,gamma):
    """
    Caculer A, F, T, alpha, beta, chi**2
    
    """
    
    R11 = intercovariance(X1,X1,N,step)
    F1 = off(R11)
    T1 = tr(R11)
    #print(F1)
    #print(T1)
    
    R22 = intercovariance(X2,X2,N,step)
    F2 = off(R22)
    T2 = tr(R22)
    #print(F2)
    #print(T2)
    
    
    R12 = intercovariance(X1,X2,N,step)
    F12 = off(R12)
    T12 = tr(R12)
    
    
    alpha = 2*F12*T12 - (F1*(T2 -gamma**2) + F2*(T1 -gamma**2))
    
    beta = 2*(T12**2 - (T1 - gamma**2)*(T2 - gamma**2))
    
    chi2 = (F1*(T2-gamma**2)-F2*(T1-gamma**2))**2 + 4*(F12*(T2-gamma*2)-T12*F2)*(F12*(T1-gamma**2)-T12*F1)
    
    d1 = alpha - chi2**(1/2)
    
    d2 = alpha + chi2**(1/2)
    
    A11 = beta*F1 - (T1 - gamma**2)*d1
    A12 = beta*F12 - T12*d2
    A21 = beta*F12 -T12*d1
    A22 = beta*F2 - (T2 -gamma**2)*d2
    
    A = np.array([[A11, A12],[A21, A22]])
    
    return A
```
